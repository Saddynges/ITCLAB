```
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

print("Upload your Excel file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

sheet_name = 'Sheet1'
data = pd.read_excel(file_name, sheet_name=sheet_name)

print("Here's the data from your file:")
print(data.head())

def analyze_feedback(feedback):
    positive_words = ["excellent", "great", "good", "amazing", "satisfied", "fast", "nice"]
    negative_words = ["bad", "poor", "disappointed", "unhappy", "terrible", "unsatisfied", "improve"]
    feedback = str(feedback).lower()
    if any(word in feedback for word in positive_words):
        return "Positive"
    elif any(word in feedback for word in negative_words):
        return "Negative"
    else:
        return "Neutral"

data['Checking'] = data['Feedback'].apply(analyze_feedback)

print("Here's the updated data with feedback types:")
print(data.head())

feedback_summary = data['Checking'].value_counts()
plt.figure(figsize=(8, 6))
feedback_summary.plot(kind='bar', color=['green', 'red', 'blue'], alpha=0.7)
plt.title("Feedback Analysis")
plt.xlabel("Feedback Type")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()

chart_file_name = 'Feedback_Analysis_Chart.png'
plt.savefig(chart_file_name)
plt.show()

updated_file_name = 'Updated_Excel_Project.xlsx'
data.to_excel(updated_file_name, index=False)

wb = load_workbook(updated_file_name)
ws = wb.active
img = Image(chart_file_name)
img.anchor = 'G1'
ws.add_image(img)
wb.save(updated_file_name)

files.download(updated_file_name)

print("Your updated Excel file with the chart is ready for download!")
```
